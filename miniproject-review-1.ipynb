The objective of this project is to develop an accurate and real-time credit card fraud detection system. The challenge lies in effectively identifying potentially fraudulent transactions in a highly imbalanced dataset while ensuring low latency processing to prevent unauthorized transactions. 

<h2>Problem Description</h2>
Credit card fraud is a wide-ranging term for theft and fraud committed using or involving a payment card, such as a credit card or debit card, as a fraudulent source of funds in a transaction. The purpose may be to obtain goods without paying or to obtain unauthorized funds from an account. Credit card fraud is also an adjunct to identity theft.
Although incidences of credit card fraud are limited to about 0.1% of all card transactions, they have resulted in huge financial losses as the fraudulent transactions have been large value transactions. It is important that credit card companies are able to recognize fraudulent credit card transactions so that customers are not charged for items that they did not purchase. What we need is an algorithm, which could classify a transaction as fraudulent or non-fraudulent. Doing so will benefit both the credit card companies and the customers who have to go through the ordeal.

## Information Gain Method

The Information Gain method is a commonly used technique in the field of machine learning and decision tree-based algorithms. It is employed to determine the relevance of a feature in a dataset when constructing decision trees or making decisions based on data.

### Formula

The Information Gain (IG) is calculated using the formula:

IG(D, A) = H(D) - H(D|A) 

Where:
- \( IG(D, A) \) represents the Information Gain of feature \( A \) in dataset \( D \).
- \( H(D) \) is the entropy of the dataset \( D \) before considering feature \( A \).
- \( H(D|A) \) is the conditional entropy of the dataset \( D \) after considering feature \( A \).

### Description

The Information Gain method measures the reduction in uncertainty in a dataset \( D \) when a particular feature \( A \) is used for classification or partitioning. It quantifies how much knowing the feature \( A \) reduces the uncertainty (or entropy) in the dataset. The higher the Information Gain, the more useful the feature is for decision-making.

- *Entropy (H(D))*: It is a measure of the impurity or disorder in the dataset before considering feature \( A \). The formula for entropy is:
     
  

  Where \( c \) is the number of classes, and \( p_i \) is the proportion of data points belonging to class \( i \) in dataset \( D \).

- *Conditional Entropy (H(D|A))*: It is a measure of the uncertainty in the dataset \( D \) after considering feature \( A \). The formula for conditional entropy is similar to entropy but weighted by the probability of the feature values:

 IG =H(D) - ∑[ (|Di| / |D|) * Entropy(H(D)i) ]

  Where \( v \) is the number of possible values for feature \( A \), \( |D_i| \) is the number of data points with the \( i \)th value of feature \( A \), and \( |D| \) is the total number of data points in dataset \( D \).

In summary, the Information Gain method calculates the difference between the entropy of the original dataset \( D \) and the weighted sum of entropies of subsets created by splitting the data based on feature \( A \). Features with higher Information Gain are preferred for building decision trees as they provide more information for accurate classification or predictions.

```python
X = df.drop(columns=["Class"]).to_numpy()  
Y = df["Class"].to_numpy()
def entropy(y):
    unique, counts = np.unique(y, return_counts=True)
    probabilities = counts / len(y)
    return -np.sum(probabilities * np.log2(probabilities))
def information_gain(X, y, feature_index):
    total_entropy = entropy(y)
    unique_values = np.unique(X[:, feature_index])
    weighted_entropy = 0

    for value in unique_values:
        subset_y = y[X[:, feature_index] == value]
        weighted_entropy += (len(subset_y) / len(y)) * entropy(subset_y)
    res= total_entropy - weighted_entropy
    print(res)
    return res;
num_features = X.shape[1]
information_gains = [information_gain(X, Y, i) for i in range(num_features)]
for i, gain in enumerate(information_gains):
    print(f"Feature {i}: Information Gain = {gain}")
top_feature_indices = np.argsort(information_gains)[::-1][:6]
print("Top 6 Features:")
for i, feature_index in enumerate(top_feature_indices):
    print(f"Feature {i + 1}: Information Gain = {information_gains[feature_index]}, Index = {feature_index}")
```

<h1 style="color : blue">Process 3: Bayesian Optimization</h1>

Bayesian Optimization is a global optimization technique used to find the maximum or minimum of a function. It is especially effective for functions that are expensive to evaluate or have a complex landscape.

The main idea behind Bayesian Optimization is to model the objective function and make intelligent decisions about where to sample next. It uses a surrogate model (usually a Gaussian Process) to approximate the true objective function.

### Key Steps in Bayesian Optimization:

1. **Define the Search Space**: Specify the range of parameter values that the optimization algorithm can explore.

2. **Define the Objective Function**: This is the function that needs to be optimized. It could be anything from machine learning model hyperparameters to tuning complex simulations.
   
3. **Select the Optimal Result**: The final result is the point that the acquisition function suggests has the highest probability of being the global maximum.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score

pbounds = {
    'learning_rate': (0.01, 0.1),  
    'max_iter': (300, 3000)  
}


def objective_function(learning_rate, max_iter):
    
    logistic_model = LogisticRegression(max_iter=int(max_iter), random_state=42)
    

    cv = StratifiedKFold(n_splits=5) 
    
  
    scores = cross_val_score(logistic_model, X_train, y_train, cv=cv, scoring='neg_log_loss')
    
    
    return -np.mean(scores)


optimizer = BayesianOptimization(
    f=objective_function,
    pbounds=pbounds,
    random_state=42,
)


optimizer.maximize(
    init_points=3, 
    n_iter=7, 
)


optimal_learning_rate = optimizer.max['params']['learning_rate']
optimal_max_iter = optimizer.max['params']['max_iter']
print(f"Optimal Learning Rate: {optimal_learning_rate}")
print(f"Optimal Number of Iterations: {optimal_max_iter}")
```

<h3>Function 1:Activation Function - sigmoid(z)</h3><br>
Description: <em>The sigmoid function squashes the output to a range between 0 and 1.</em><br>

Mathematical Explanation:<br>
<strong>  $g(z) = \frac{1}{1+e^{-z}}\tag{3}$</strong>

epsilon = 1e-10:

epsilon is a small constant added to the sigmoid function to avoid numerical instability issues. When the input to the sigmoid function is too large (in magnitude), it can lead to numerical overflow. By adding a small constant like epsilon, it ensures that the value passed to the exp function is never too large.


```python
epsilon = 1e-10
def sigmoid(z):
    return 1 / (1 + np.exp(-z + epsilon))
```

<h3>Function 2: initialize_weights(num_features)</h3><br>
Description: <em>Initializes the weights and bias to zero.</em><br>
<a href="https://datascience.stackexchange.com/questions/30989/what-are-the-cases-where-it-is-fine-to-initialize-all-weights-to-zero#:~:text=Whenever%20you%20have%20a%20convex,cross%2Dentropy%20for%20logistic%20regression" style="color:red" > <b>Reference Link for zero intialization</b></a>

```python
def initialize_weights(num_features):
    w = np.zeros(num_features)
    b = 0
    return w, b
```

## Cost function

For logistic regression, the cost function is of the form 

$$ J(\mathbf{w},b) = \frac{1}{m} \sum_{i=0}^{m-1} \left[ loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), y^{(i)}) \right] \tag{1}$$

where
* $loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), y^{(i)})$ is the cost for a single data point, which is:

    $$loss(f_{\mathbf{w},b}(\mathbf{x}^{(i)}), y^{(i)}) = -y^{(i)} \log\left(f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) - \left( 1 - y^{(i)}\right) \log \left( 1 - f_{\mathbf{w},b}\left( \mathbf{x}^{(i)} \right) \right) \tag{2}$$
    
*  where m is the number of training examples in the data set and:
$$
\begin{align}
  f_{\mathbf{w},b}(\mathbf{x^{(i)}}) &= g(z^{(i)})\tag{3} \\
  z^{(i)} &= \mathbf{w} \cdot \mathbf{x}^{(i)}+ b\tag{4} \\
  g(z^{(i)}) &= \frac{1}{1+e^{-z^{(i)}}}\tag{5} 
\end{align}
$$

<h3>Function 3: compute_cost(X, y, w, b)</h3><br>
Description: <em>Computes the cost function, which measures how well the model predicts the target variable.</em><br>

Mathematical Explanation:<br>
<strong>J(w) = - (1/m) Σ [ y*log(h) + (1-y)*log(1-h) ]</strong>
Where:
<ul>
    <li>J(w) is the cost function.</li>
    <li>m is the number of training examples.</li>
    <li>y(i)is the true label of the i-th training example.</li>
    <li>h(i)is the predicted probability that y(i)=1</li>
    <li>The summation is taken over all training examples.</li>
    </ul>

```python
def compute_cost(X, y, w, b, class_weights=None):
    m = len(y)
    z = np.dot(X, w) + b
    h = sigmoid(z)
    
    if class_weights is not None:
        cost = -(1/m) * np.sum(y*np.log(h + epsilon) * class_weights + (1-y)*np.log(1-h + epsilon))
    else:
        cost = -(1/m) * np.sum(y*np.log(h + epsilon) + (1-y)*np.log(1-h + epsilon))
        
    return cost
```

## Logistic Gradient Descent


$$\begin{align*}
&\text{repeat until convergence:} \; \lbrace \\
&  \; \; \;w_j = w_j -  \alpha \frac{\partial J(\mathbf{w},b)}{\partial w_j} \tag{1}  \; & \text{for j := 0..n-1} \\ 
&  \; \; \;  \; \;b = b -  \alpha \frac{\partial J(\mathbf{w},b)}{\partial b} \\
&\rbrace
\end{align*}$$

Where each iteration performs simultaneous updates on $w_j$ for all $j$, where
$$\begin{align*}
\frac{\partial J(\mathbf{w},b)}{\partial w_j}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)})x_{j}^{(i)} \tag{2} \\
\frac{\partial J(\mathbf{w},b)}{\partial b}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}) \tag{3} 
\end{align*}$$

* m is the number of training examples in the data set      
* $f_{\mathbf{w},b}(x^{(i)})$ is the model's prediction, while $y^{(i)}$ is the target
* For a logistic regression model  
    $z = \mathbf{w} \cdot \mathbf{x} + b$  
    $f_{\mathbf{w},b}(x) = g(z)$  
    where $g(z)$ is the sigmoid function:  
    $g(z) = \frac{1}{1+e^{-z}}$   
    
    

<h3>Function 4: gradient_descent(X, y, w, b, learning_rate)</h3><br>
Description: <em>Computes the cost function, which measures how well the model predicts the target variable.</em><br>

Mathematical Explanation:<br>
<strong>w_j = w_j - α * (1/m) Σ (h^(i) - y^(i)) * x_j^(i)</strong><br>

<strong>b = b - α * (1/m) Σ (h^(i) - y^(i))</strong>
Where:
<ul>
    <li>w_j represents the j-th weight.</li>
    <li>α is the learning rate.</li>
    <li>m is the number of training examples.</li>
    <li>y(i)is the true label of the i-th training example.</li>
    <li>h(i)is the predicted probability that y(i)=1</li>
    <li>x_j^(i)is the j-th feature of the i-th example.</li>
    </ul>

```python
def gradient_descent(X, y, w, b, learning_rate, class_weights=None):
    m = len(y)
    z = np.dot(X, w) + b
    h = sigmoid(z)
    
    if class_weights is not None:
        dw = (1/m) * np.dot(X.T, (h - y) * class_weights[y.astype(int)])
    else:
        dw = (1/m) * np.dot(X.T, (h - y))
    
    db = (1/m) * np.sum(h - y)
    
    w = w - learning_rate * dw
    b = b - learning_rate * db
    
    return w, b
```

<h3>Function 5: train_logistic_regression(X, y, num_iterations, learning_rate)</h3><br>
Description: <em>Orchestrates the training process of the logistic regression model.</em><br>

```python
def train_logistic_regression(X, y, num_iterations, learning_rate, class_weights=None):
    w, b = initialize_weights(X.shape[1])
    costs = []  
    
    for i in range(num_iterations):
        cost = compute_cost(X, y, w, b)
        costs.append(cost) 
        
        if class_weights is not None:
            w, b = gradient_descent(X, y, w, b, learning_rate, class_weights)
        else:
            w, b = gradient_descent(X, y, w, b, learning_rate)
    
   
    plt.plot(range(num_iterations), costs)
    plt.xlabel('Iterations')
    plt.ylabel('Cost')
    plt.title('Gradient Descent Convergence')
    plt.show()
    
    return w, b
```

<h3>Function 6: compute_class_weights(y)</h3><br>
Description: <em>This function computes the class weights for handling imbalanced datasets in logistic regression.</em><br>

Mathematical Explanation:<br>
$w_j= \frac{{samples}}{samples_j * classes}$   

<strong>w_j = n_samples / ( n_samples_j * n_classes)
</strong><br>
Where:
<ul>
    <li>w_jrepresents the weight for class j</li>
    <li>n_samples is the total number of samples or rows in the dataset.</li>
    <li>n_classes is the total number of unique classes in the target variable.</li>
    <li>n_samples_j is the total number of rows belonging to class j.</li>
    </ul>
    
The class weight w_j is calculated to give more importance to the minority class during training. 

```python
def compute_class_weights(y):
    n_samples = len(y)
    n_classes = len(np.unique(y))
    n_samples_j = (np.bincount(y))
    print(f"Total number of samples: {n_samples}")
    print(f"Total number of classes: {n_classes}")
    print(f"Total number of samples belonging to Non-Fraudulent : {n_samples_j[0]}")
    print(f"Total number of samples belonging to Fraudulent : {n_samples_j[1]}")
    class_weights = n_samples / (n_samples_j * n_classes )
    return class_weights

class_weights = compute_class_weights(y_train).ravel()
print(f"Class weight for non fraudulent is {class_weights[0]}")
print(f"Class weight for Fraudulent is {class_weights[1]}")
```

<h2 style="color: red">Step 5: Evaluate Model Performance</h2>

  $$f_{\mathbf{w},b}(\mathbf{x}^{(i)}) = g(\mathbf{w} \cdot \mathbf{x}^{(i)} + b) \tag{1}$$

  where $g(z)$ is known as the sigmoid function and it maps all input values to values between 0 and 1:

  $g(z) = \frac{1}{1+e^{-z}}\tag{2}$
  and $\mathbf{w} \cdot \mathbf{x}$ is the vector dot product:
  
  $$\mathbf{w} \cdot \mathbf{x} = w_0 x_0 + w_1 x_1$$
  
  
 * We interpret the output of the model ($f_{\mathbf{w},b}(x)$) as the probability that $y=1$ given $\mathbf{x}$ and parameterized by $\mathbf{w}$ and $b$.
* Therefore, to get a final prediction ($y=0$ or $y=1$) from the logistic regression model, we can use the following heuristic -

  if $f_{\mathbf{w},b}(x) >= 0.5$, predict $y=1$
  
  if $f_{\mathbf{w},b}(x) < 0.5$, predict $y=0$
  

```python
def evaluate_model(X, y, w, b): 
    z = np.dot(X, w) + b
    h = sigmoid(z)
    y_pred = (h >= 0.5).astype(int) 
    accuracy = accuracy_score(y, y_pred)
    precision = precision_score(y, y_pred)
    recall = recall_score(y, y_pred)
    f1 = f1_score(y, y_pred)
    roc_auc = roc_auc_score(y, h)
    mcc = matthews_corrcoef(y, y_pred)
    return accuracy, precision, recall, f1, roc_auc,mcc,y_pred
```

# Five-Fold Cross Validation with Logistic Regression

In this section, we perform five-fold cross-validation using logistic regression to evaluate the model's performance.

### Methodology

1. **Dataset Splitting**: The dataset is divided into five approximately equal parts.

2. **Model Training and Evaluation**:
   - For each fold:
     - One part is used as the validation set, while the remaining four parts are used for training.
     - The logistic regression model is trained on the training set.
     - The trained model is then evaluated on the validation set.

3. **Metrics Calculation**: Various performance metrics are calculated for each fold.

   - **F1 Score**: The F1 Score is the harmonic mean of precision and recall. It provides a balance between false positives and false negatives.
   - **Precision**: Precision is the ratio of true positives to the sum of true positives and false positives. It measures the accuracy of positive predictions.
   - **Recall**: Recall is the ratio of true positives to the sum of true positives and false negatives. It measures the ability to correctly identify positive instances.
   - **Accuracy**: Accuracy is the ratio of correct predictions to the total number of predictions. It measures overall correctness.
   - **ROC AUC**: ROC AUC (Receiver Operating Characteristic Area Under the Curve) measures the area under the ROC curve, which represents the model's ability to distinguish between classes.
   - **Matthews Correlation Coefficient (MCC)**: MCC is a correlation coefficient between observed and predicted binary classifications. It ranges from -1 to 1, where 1 indicates perfect prediction.

4. **Confusion Matrices**: Additionally, a confusion matrix is generated for each fold to provide a detailed breakdown of model predictions.



  ### Function Description

  This function performs five-fold cross-validation using `StratifiedKFold` with `n_splits=5`. It evaluates logistic regression models using different subsets of the data for training and validation. The evaluation metrics like F1 score, precision, recall, accuracy, ROC AUC, and MCC are computed for each fold and then averaged.

  ### Explanation of Cross-Validation

  `StratifiedKFold` is a technique for splitting the dataset into `k` consecutive folds while ensuring that each fold maintains the same proportion of target classes as the original dataset. In this case, `n_splits=5` means the data is divided into 5 subsets for cross-validation.
  n.

After completing the five-fold cross-validation process, the following average metrics are obtained:

- **Average Accuracy**: {:.4f}

```python
class DecisionTree:
    def __init__(self, max_depth=3):
        self.max_depth = max_depth
        self.tree = None

    def fit(self, X, gradients, hessians, depth=0):
        if depth == self.max_depth or len(set(gradients)) == 1:
            self.tree = sum(gradients) / (sum(hessians) + 1e-8)
            return

        best_split = None
        best_gain = float('-inf')
        for feature_idx in range(len(X[0])):
            thresholds = set(x[feature_idx] for x in X)
            for threshold in thresholds:
                left_idx = [i for i, x in enumerate(X) if x[feature_idx] <= threshold]
                right_idx = [i for i, x in enumerate(X) if x[feature_idx] > threshold]

                if not left_idx or not right_idx:
                    continue

                gain = self.calculate_gain(gradients, hessians, left_idx, right_idx)
                if gain > best_gain:
                    best_gain = gain
                    best_split = (feature_idx, threshold, left_idx, right_idx)

        if best_split is None:
            self.tree = sum(gradients) / (sum(hessians) + 1e-8)
            return

        _, _, left_idx, right_idx = best_split
        left_X = [X[i] for i in left_idx]
        right_X = [X[i] for i in right_idx]

        left_gradients = [gradients[i] for i in left_idx]
        left_hessians = [hessians[i] for i in left_idx]
        right_gradients = [gradients[i] for i in right_idx]
        right_hessians = [hessians[i] for i in right_idx]

        self.tree = [DecisionTree(self.max_depth), DecisionTree(self.max_depth), best_split]
        self.tree[0].fit(left_X, left_gradients, left_hessians, depth + 1)
        self.tree[1].fit(right_X, right_gradients, right_hessians, depth + 1)

    def calculate_gain(self, gradients, hessians, left_idx, right_idx):
        left_gradients = sum(gradients[i] for i in left_idx)
        left_hessians = sum(hessians[i] for i in left_idx)
        right_gradients = sum(gradients[i] for i in right_idx)
        right_hessians = sum(hessians[i] for i in right_idx)
        total_gradients = left_gradients + right_gradients
        total_hessians = left_hessians + right_hessians

        gain = (left_gradients ** 2 / (left_hessians + 1e-8)) + \
               (right_gradients ** 2 / (right_hessians + 1e-8)) - \
               (total_gradients ** 2 / (total_hessians + 1e-8))
        return gain

    def predict(self, X):
        if not self.tree or isinstance(self.tree, float):
            return self.tree

        feature_idx, threshold, _, _ = self.tree[2]
        if X[feature_idx] <= threshold:
            return self.tree[0].predict(X)
        else:
            return self.tree[1].predict(X)

    def print_tree(self, depth=0, feature_names=None):
        if isinstance(self.tree, float):
            print("\t" * depth + f"Leaf: {self.tree}")
        else:
            feature_name = feature_names[self.tree[2][0]] if feature_names else f"Feature {self.tree[2][0]}"
            threshold = self.tree[2][1]
            print("\t" * depth + f"{feature_name} <= {threshold}")
            self.tree[0].print_tree(depth + 1, feature_names)
            print("\t" * depth + f"{feature_name} > {threshold}")
            self.tree[1].print_tree(depth + 1, feature_names)
```

<h1 style="font-size:48px;">Model 2: LightGBM</h1>

<p>LightGBM is a gradient-boosting framework based on decision trees to increase the efficiency of the model and reduces memory usage.</p>
<h5>LightGBM grows trees leaf-wise (best-first). It will choose the leaf with max delta loss to grow. Holding #leaf fixed, leaf-wise algorithms tend to achieve lower loss than level-wise algorithms.

Leaf-wise may cause over-fitting when #data is small, so LightGBM includes the max_depth parameter to limit tree depth. However, trees still grow leaf-wise even when max_depth is specified.</h5>



<h3>Introduction to the TreeNode Class</h3>
<p>
  The <code>TreeNode</code> class is fundamental to the LightGBM's leaf-wise tree growth strategy. This class encapsulates the structure and behavior of each node in the decision tree, including aspects like depth, threshold for splitting, and the left and right child nodes.
</p>
<h3>Fit Method Explanation</h3>
<p>
  The <code>fit</code> method in the <code>TreeNode</code> class is responsible for training the decision tree. It involves finding the best split based on Gini impurity and recursively growing left and right child trees. This method embodies the leaf-wise growth strategy central to LightGBM.
</p>
<h3>Finding the Best Split</h3>
<p>
  The <code>_find_best_split</code> method iterates over all features and their possible thresholds to determine the most optimal split. It aims to minimize the Gini impurity, which is a measure of how often a randomly chosen element would be incorrectly identified.
</p>
<h3>Gini Impurity Calculation</h3>
<p>
  The <code>_calculate_gini</code> method calculates the Gini impurity for a set of labels. This measure helps in evaluating the quality of the splits during the tree growth process.
</p>
<h3>Visualizing the Tree</h3>
<p>
  The <code>plot_tree</code> method provides a visual representation of the decision tree. This visualization is crucial for understanding the tree structure and the decision paths in the tree.
</p>
<h3>Leaf-Wise vs Level-Wise Tree Growth</h3>
<p>
  LightGBM's unique approach involves splitting the tree leaf-wise rather than level-wise. This method selects the leaf with the maximum delta loss for growth, leading to a more efficient and accurate model, especially on large datasets. However, it's important to note that leaf-wise growth can potentially lead to overfitting in cases of small datasets.
</p>

```python
class TreeNode:
    def __init__(self, depth, max_depth):
        self.depth = depth
        self.max_depth = max_depth
        self.feature_index = None
        self.threshold = None
        self.value = None
        self.left = None
        self.right = None

    def fit(self, X, y):
        if self.depth == self.max_depth or len(set(y)) == 1:
            self.value = np.mean(y)
        else:
            self.feature_index, self.threshold = self._find_best_split(X, y)
            print("self.feature_index, self.threshold")
            print(self.feature_index, self.threshold)
            left_mask = X[self.feature_index] <= self.threshold
            right_mask = ~left_mask
            print("left_mask")
            print(left_mask)
            print("right_mask")
            print(right_mask)

            self.left = TreeNode(self.depth + 1, self.max_depth)
            self.right = TreeNode(self.depth + 1, self.max_depth)
            #print("X.loc[left_mask]")
            #print(X.loc[left_mask])
            #print("y[left_mask]")
            #print(y[left_mask])
            self.left.fit(X.loc[left_mask], y[left_mask])
            self.right.fit(X.loc[right_mask], y[right_mask])

    def _find_best_split(self, X, y):
        m, n = X.shape
        print(m)
        print(n)
        best_gini = float('inf')
        best_feature, best_threshold = None, None
        
        for feature_name in X.columns:
            thresholds = np.unique(X[feature_name])
            
            for threshold in thresholds:
                left_mask = X[feature_name] <= threshold
                #print("left_mask")
                #print(left_mask)
                right_mask = ~left_mask
                #print("right_mask")
                #print(right_mask)
                
                gini = self._calculate_gini(y[left_mask]) * np.sum(left_mask) / m
                
                #print("calculate_gini",self._calculate_gini(y[left_mask])) 
                #print("gini") 
                #print(gini) 
                
                gini += self._calculate_gini(y[right_mask]) * np.sum(right_mask) / m

                if gini < best_gini:
                    best_gini = gini
                    best_feature = feature_name
                    best_threshold = threshold

        return best_feature, best_threshold

    def _calculate_gini(self, y):
        #print('y value')
        #print(y)
        if len(y) == 0:
            return 0
        p = np.bincount(y) / len(y)
        #print("P value")
        #print(np.bincount(y))
        #print(p)
        return 1 - np.sum(p ** 2)
    
    def plot_tree(self, ax, x_offset=0, y_offset=0, x_spacing=1, y_spacing=1):
        if self.value is not None:
            ax.text(x_offset, y_offset, f"{self.value:.2f}", ha='center', va='center', bbox=dict(facecolor='lightgray', edgecolor='gray'))
        else:
            ax.text(x_offset, y_offset, f"X{self.feature_index} <= {self.threshold:.2f}", ha='center', va='center', bbox=dict(facecolor='lightgray', edgecolor='gray'))
            left_x = x_offset - x_spacing
            right_x = x_offset + x_spacing
            y = y_offset - y_spacing
            ax.plot([x_offset, left_x], [y, y - y_spacing], color='black')
            ax.plot([x_offset, right_x], [y, y - y_spacing], color='black')
            self.left.plot_tree(ax, left_x, y - y_spacing, x_spacing, y_spacing)
            self.right.plot_tree(ax, right_x, y - y_spacing, x_spacing, y_spacing)

```

```python
from sklearn.metrics import make_scorer, matthews_corrcoef
from sklearn.model_selection import cross_validate, StratifiedKFold
import pandas as pd
from tabulate import tabulate
import numpy as np

def cv_results(model, X, y, model_name, output_type='result'):
    sk_fold = StratifiedKFold(n_splits=5)

    scoring = {
        'f1': 'f1',
        'precision': 'precision',
        'recall': 'recall',
        'accuracy': 'accuracy',
        'roc_auc': 'roc_auc',
        'mcc': make_scorer(matthews_corrcoef)
    }

    cv_scores = cross_validate(model, X, y, scoring=scoring, cv=sk_fold)

    avg_scores = {
        'f1': round(np.mean(cv_scores['test_f1']), 8),
        'precision': round(np.mean(cv_scores['test_precision']), 8),
        'recall': round(np.mean(cv_scores['test_recall']), 8),
        'accuracy': round(np.mean(cv_scores['test_accuracy']), 8),
        'roc_auc': round(np.mean(cv_scores['test_roc_auc']), 8),
        'mcc': round(np.mean(cv_scores['test_mcc']), 8)
    }

    if output_type == 'dict':
        return avg_scores
    elif output_type == 'result':
        return cv_scores
    elif output_type == 'print':
        # Create a DataFrame with these metrics
        metrics_df = pd.DataFrame([avg_scores])

        # Insert the model name as the first column
        metrics_df.insert(0, 'Model Name', model_name)

        # Use tabulate to print the DataFrame with borders in horizontal format
        print(tabulate(metrics_df, headers='keys', tablefmt='pretty', showindex=False))
```
